# Support Vector Machine (SVM)

## Motivation:
Classfication: Finding w with large margin

Let $x_n$ be the nearest data point to the plane
$w^Tx + b = 0$.

What is the distance?
The distance between $x_n$ and the plane. The vector w is orthogonal to the plane in the X space.
## Distance
### Functional Distance
$\hat{\gamma} = y(w^Tx+b)=yf(x)$
$\hat{\gamma}=min \hat{\gamma_i},$(i=1,..,n)


### Geometric Distance

Suppose there is a point x, the project of x on the plane is x0. w is a vector orthogona to the plane, $\gamma$ is the distance between x to the plane. so

$x = x_0 + \gamma \frac{w}{||w||}$, since $f(x_0)=0$, so 
\begin{align*}
f(x) &= f(x_0) + f(\gamma \frac{w}{||w||})-b\\
&=0+w^T \gamma \frac{w}{||w||}+b-b\\
&=\gamma \frac{w^T*w}{||w||}\\
&=\gamma \frac{||w||^2}{||w||}\\
&=\gamma ||w||\\
\therefore \gamma &= \frac{f(x)}{||w||}
\end{align*}

To get the absolute value,
$\tilde{\gamma}=y\gamma=\frac{\hat{\gamma}}{||w||}$

So our goal is to:
## Optimization Problem
\begin{align}
max [\tilde{\gamma}]=\frac{1}{||w||},s.t. y_i(w^Tx_i+b)>=1,i=1,...,n
\end{align}

This is equivalent to:

\begin{align}
min 1/2 ||w||^2,
s.t. y_i(w^Tx_i+b)>=1,i=1,...,n
\end{align}

This is Quadratic Programming problem. We can solve it with Lagrange Duality to solve it.

## Lagrange Duality

$L(w,b,\alpha)=1/2||w||^2-\sum_{i=1}^n \alpha_i(y_i(w^Tx_i+b)-1)$
solve:
$min_{w,b}max_{\alpha_i>=0} L(w,b,\alpha)=p^*$

this is similar to:

$max_{\alpha_i>=0}min_{w,b} L(w,b,\alpha)=d^*$ and $d^*<= p^*$ under K.K.T condition.

### K.K.T condition

\begin{align*}
\text{min f(x)}\\
s.t. h_j(x)=0\\
g_k(x)<=0
\end{align*}

1.$h_j(x^*)=0, g_k(x^*)<=0$

2.$\triangle f(x^*)+\sum_{j=1}^p \lambda_j \triangle h_j(x^*) +\sum_{k=1}^q \mu_k \triangle g_k(x^*)=0$

## Solve duality in 3 steps
### 1 fix $\alpha$, find minimum L with w and b
$\frac{\delta L}{\delta w}=0 \to w = \sum_{i=1}^n \alpha_iy_ix_i$

$\frac{\delta L}{\delta b}=0 \to \sum_{i=1}^n \alpha_i y_i=0$

\begin{align*}
L(w,b,\alpha) &= 1/2 ||w||^2-\sum_{i=1}^n \alpha_i(y_i(w^Tx_i+b)-1)\\
&= 1/2 w^T w - \sum_{i=1}^n \alpha_iy_iw^Tx_i - \sum_{i=1}^n \alpha_iy_ib + \sum_{i=1}^n \alpha_i\\
&=1/2w^T\sum_{i=1}^n\alpha_iy_ix_i - \sum_{i=1}^n \alpha_iy_iw^Tx_i - \sum_{i=1}^n \alpha_iy_ib + \sum_{i=1}^n \alpha_i\\
&=1/2w^T\sum_{i=1}^n\alpha_iy_ix_i - w^T\sum_{i=1}^n \alpha_iy_ix_i - \sum_{i=1}^n \alpha_iy_ib + \sum_{i=1}^n \alpha_i\\
&=-1/2w^T\sum_{i=1}^n\alpha_iy_ix_i - \sum_{i=1}^n \alpha_iy_ib + \sum_{i=1}^n \alpha_i\\
&=-1/2w^T\sum_{i=1}^n\alpha_iy_ix_i - b\sum_{i=1}^n \alpha_iy_i + \sum_{i=1}^n \alpha_i\\
&=-1/2(\sum_{i=1}^n\alpha_iy_ix_i)^T\sum_{i=1}^n\alpha_iy_ix_i - b\sum_{i=1}^n \alpha_iy_i + \sum_{i=1}^n \alpha_i\\
&=-1/2\sum_{i=1}^n\alpha_iy_i(x_i)^T\sum_{i=1}^n\alpha_iy_ix_i - b\sum_{i=1}^n \alpha_iy_i + \sum_{i=1}^n \alpha_i\\
&=-1/2\sum_{i=1}^n\alpha_iy_i(x_i)^T\alpha_iy_ix_i - b\sum_{i=1}^n \alpha_iy_i + \sum_{i=1}^n \alpha_i\\
&=-1/2\sum_{i,j=1}^n\alpha_i\alpha_jy_iy_jx_i^Tx_j^T\alpha_iy_ix_i + \sum_{i=1}^n \alpha_i\\
\end{align*}

### 2 Max L with $\alpha$ 

$max_{\alpha} \sum_{i=1}^n\alpha_i-1/2\sum_{i,j=1}^n\alpha_i\alpha_jy_iy_jx_i^Tx_j, s.t. \alpha_i>=0,i=1,..,n, \sum_{i=1}^n\alpha_iy_i=0$

Find $\alpha_i$, and

$w^*=\sum_{i=1}^n \alpha_iy_ix_i$

$b^*=-\frac{max_{i:y_i=-1}w^{*T}x_i+ min_{i:y_i=1}w^{*T}x_i}{2}$
